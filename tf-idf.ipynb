```python
## example of tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
# Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

messages=pd.read_csv('./data/email.csv',
                    sep=',', names=['label', 'message'])
ps = PorterStemmer()
wordnet=WordNetLemmatizer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    # review = review.lower()
    # review = review.split()
    # review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    # review = ' '.join(review)
    corpus.append(review)


vectorizer = TfidfVectorizer(max_features=100, ngram_range=(2, 2))
X = vectorizer.fit_transform(corpus).toarray()

np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3f" % x))
vectorizer.vocabulary_
```
